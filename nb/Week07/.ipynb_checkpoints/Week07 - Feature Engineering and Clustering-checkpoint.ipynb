```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn 
```

```python
data = np.genfromtxt("../../data/newdata/AllChiPurged.dat")
```

```python
# Create dataframe for easier operations/manipulation
df = pd.DataFrame(data, columns = ['ustat','alph1','alph2', 'beta', 'chi'])
```

#### Feature Engineering

Let's bring back the feature 'diff' and drop the features alph1 and alph2. My reasoning for this is that now we will have exactly three features for clustering.

```python
# diff = |alph1 - alph2|
diff = []

for i in range(len(df['alph1'])):
    diff.append(abs(df['alph1'][i] - df['alph2'][i]))
    
df['diff'] = diff
```

```python
df = df.drop(['alph1', 'alph2'], axis=1)
```

```python
X_df = df.drop(['chi'], axis=1)
y_df = df[['chi']]
```

#### Standardize and k-Means Clustering

This is a repeat of the steps from the previous week. We use power trasnformer function to standardize the data and k-means to create clusters.

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')    
```

```python
# transform features
X_df = pt.fit_transform(X_df)
X_df = pd.DataFrame(X_df, columns = ['ustat','beta', 'diff'])
```

```python
X_df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
# elbow method: heuristic used in determining the number of clusters in a data set
# SSE/intertia: sum of the squared distance between the average point (called Centroid) and each point of the cluster

def elbow_method(X):
    error = []
    for i in range(1,11):
        k_means = KMeans(n_clusters=i,init='k-means++', random_state=42)
        k_means.fit(X)
        # Sum of squared distances of samples to their closest cluster center, 
        # weighted by the sample weights if provided.
        error.append(k_means.inertia_)
    #plot elbow curve
    plt.plot(np.arange(1,11), error)
    plt.xlabel('k')
    plt.ylabel('Sum of Square Error (SSE)')
    plt.show()
```

```python
# extract three features/columns
# ustat, beta, diff
X = X_df.values
```

```python
elbow_method(X)
```

```python
# predict labels for K = 3 clusters
km = KMeans(n_clusters = 3, init = 'k-means++',  random_state=42)
y = km.fit_predict(X)
```

```python
# labels
data1 = X_df[y==0]
data2 = X_df[y==1]
data3 = X_df[y==2]
```

```python
fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection ="3d")



# Data for three-dimensional scattered points
ax.scatter3D(data1['ustat'], data1['beta'], data1['diff'], c='tab:olive', label = 'Cluster 1')

ax.scatter3D(data2['ustat'],data2['beta'],data2['diff'],c ='tab:cyan', label = 'Cluster 2')

ax.scatter3D(data3['ustat'], data3['beta'], data3['diff'], c='tab:purple', label = 'Cluster 3')


plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'black', s = 100)

ax.set_title('k-Means Cluster, k = 3: ustat, beta, diff')
ax.set_xlabel('ustat')
ax.set_ylabel('beta')
ax.set_zlabel('diff')


plt.legend()
plt.show()
```

```python
y_df['KM01'] = y
```

```python
print(y_df['KM01'].value_counts())
df1 = y_df[['chi', 'KM01']]
df1.plot.scatter(x = 'KM01', y = 'chi')
plt.xlabel('k')
plt.ylabel('chi')
```

Cluster 0 and 1 have approximately the same number of instances. Cluster 2 is about one-third of Cluster 0. There is no way to create separate labels from the given clusters. 

Let's try k = 2...

```python
# predict labels for K = 2 clusters
km = KMeans(n_clusters = 2, init = 'k-means++',  random_state=42)
y = km.fit_predict(X)
```

```python
# labels
data1 = X_df[y==0]
data2 = X_df[y==1]
```

```python
fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection ="3d")



# Data for three-dimensional scattered points
ax.scatter3D(data1['ustat'], data1['beta'], data1['diff'], c='tab:olive', label = 'Cluster 1')

ax.scatter3D(data2['ustat'],data2['beta'],data2['diff'],c ='tab:cyan', label = 'Cluster 2')


plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'black', s = 100)

ax.set_title('k-Means Cluster, k = 2: ustat, beta, diff')
ax.set_xlabel('ustat')
ax.set_ylabel('beta')
ax.set_zlabel('diff')


plt.legend()
plt.show()
```

```python
y_df['KM01'] = y
```

```python
print(y_df['KM01'].value_counts())
df1 = y_df[['chi', 'KM01']]
df1.plot.scatter(x = 'KM01', y = 'chi')
plt.xlabel('k')
plt.ylabel('chi')
```

There is a large imbalance in the number of data instances for the clusters. There is no way to create separate labels from the given clusters.
