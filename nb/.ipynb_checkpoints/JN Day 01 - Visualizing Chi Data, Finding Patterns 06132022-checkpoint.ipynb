```python
# examine ustat v chi further

# extract wanted values, convert into array
ustat_chi = df.iloc[:,[0,4]].values
```

```python
# dbscan to find cluster relations

from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=1,min_samples=6)
```

```python
labels = dbscan.fit_predict(ustat_chi)
```

```python
# -1 == noise

np.unique(labels)
```

```python
plt.scatter(ustat_chi[labels== -1,0], ustat_chi[labels == -1,1], s = 10, c = 'red')
plt.scatter(ustat_chi[labels== 0,0], ustat_chi[labels == 0,1], s = 10, c = 'blue')
#plt.scatter(ustat_chi[labels== 1,0], ustat_chi[labels == 1,1], s = 10, c = 'orange')
#plt.scatter(ustat_chi[labels== 2,0], ustat_chi[labels == 2,1], s = 10, c = 'pink')
```

```python
# Previous work done by Bowman Paper

# ustat v Chi
# Marking Chi = 3 as boundary line separating two classes: small chi values and large chi values

scatter1 = df.plot.scatter(x='ustat', y='chi')
plt.axhline(y=3, color='r', linestyle='-')
```

```python
plt.scatter(ustat_chi[labels== -1,0], ustat_chi[labels == -1,1], s = 10, c = 'red')
plt.scatter(ustat_chi[labels== 0,0], ustat_chi[labels == 0,1], s = 10, c = 'blue')
plt.axhline(y=3, color='r', linestyle='-')
```
