```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn 
```

```python
data = np.genfromtxt("../../data/data.dat")
data = np.delete(data, 0, 0)
df = pd.DataFrame(data, columns = ['ustat','alph1','alph2', 'beta', 'chi'])
print(df.head())
print(df.shape[0])
```

```python
# Feature Engineering
```

```python
# CHI

# small chi = 0
# large chi = 1
labels = []
for i in df['chi']:
    if i < 3:
        labels.append(0)
    else:
        labels.append(1)
        
df['chi'] = labels
```

```python
# USTAT

# small ustat = 0
# large ustat = 1

labels = []
for i in df['ustat']:
    if i < 3:
        labels.append(0)
    else:
        labels.append(1)
        
df['ustat'] = labels
```

```python
# ALPH1

# small alph1 = 0
# medium alph1 = 1
# large alph1 = 2

labels = []
for i in df['alph1']:
    if i < 1:
        labels.append(0)
    elif i >= 1 and i < 2:
        labels.append(1)
    else:
        labels.append(2)
        
df['alph1'] = labels
```

```python
# ALPH2

# small alph2 = 0
# large alph2 = 1

labels = []
for i in df['alph2']:
    if i < 2:
        labels.append(0)
    else:
        labels.append(1)
        
df['alph2'] = labels
```

```python
# BETA

# small beta = 0
# large beta = 1
labels = []
for i in df['beta']:
    if i < 50:
        labels.append(0)
    else:
        labels.append(1)
        
df['beta'] = labels
```
